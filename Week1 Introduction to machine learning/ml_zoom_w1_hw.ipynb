```python
islands_housing = housing[housing['ocean_proximity']=='ISLAND']
islands_housing
```

```python
X = islands_housing[['housing_median_age','total_rooms','total_bedrooms']]
X = X.to_numpy()
```

```python
XTX = np.matmul(X.T,X)
XTX
```

```python
y = np.array([950, 1300, 800, 1000, 1300])
```

```python
XTX_inv = np.linalg.inv(XTX)
```

```python
A = np.matmul(XTX_inv,X.T)
```

```python
w = np.dot(A,y)
w
```
